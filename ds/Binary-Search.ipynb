```python
arr = [1,3,5,7,11,16,20]
target = 11


def BinarySearch(i,j):
    print(i,j)
    if j>=i:
        mid=i+(j-i)//2
        print(mid)
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            return BinarySearch(mid+1,j)
        else:
            return BinarySearch(i,mid-1)
    
    else:
        return False
    
BinarySearch(0,len(arr)-1)


    
    
    
```
