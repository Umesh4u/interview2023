````python
1. asgi and wsgi file full form and use and if delete what is impact on project
2. concurrency and parallelism
3. diamond problem
4. mro
5. mixins
6. modelviewset and apiview
7. monkeypatching ka real project example
8. what is interface
9. Duck Typing in python
10. model inheritance?
11. what is python architecure?
12. what is django architecure?
13. what is difference in json and dictionary?
14. how to make thread?
15. does python support multi programming
16. which type of data type support key in dictionary?
17. annotate & aggregate in django?
ans. In Django, annotate and aggregate are query functions that allow you to perform calculations 
and aggregations on your database queries.
annotate adds new fields to each object in the queryset based on some calculations or aggregations. 
These fields are not stored in the database but are calculated on-the-fly when the query is executed.
Here's an example of using annotate:
from django.db.models import Count
from myapp.models import MyModel

# Annotate the queryset with the count of related objects
queryset = MyModel.objects.annotate(num_related=Count('related_objects'))

# Access the annotated field for each object
for obj in queryset:
    print(obj.num_related)

aggregate, on the other hand, performs calculations across the entire queryset and returns a single 
result. It allows you to obtain aggregated values such as sums, averages, counts, etc.
Here's an example of using aggregate:
from django.db.models import Sum
from myapp.models import MyModel

# Aggregate the total value of a field across the queryset
result = MyModel.objects.aggregate(total_value=Sum('value'))

# Access the aggregated result
print(result['total_value'])



18. inline function like lambda banane ka benifit kya hai?
19. caching kya hoti hai.
ans. Caching refers to the technique of storing the output results when they are processed initially
so that next time when the same results are fetched again, instead of processing again those already 
stored results can be used, which leads to faster accessing as well us less resource utilization. 
Django provides us with a robust cache system that is able to store dynamic web pages so that these
pages don’t need to be evaluated again for each request. 
Memcached - A memory-based cache server is the fastest and most efficient
FileSystem Caching - Values of the cache are stored as separate files in a serialized order
Local-memory Caching - This is used as the default cache strategy by Django if you haven’t set anything. It is per-process as well as thread-safe.
Database Caching  - Cache data will be stored in the database and works very well if you have a fast
and well-indexed DB server.

20. gunicorn?
ans. Gunicorn (Green Unicorn) is a popular Python Web Server Gateway Interface (WSGI) HTTP server. 
It is designed to be a lightweight and simple server for running Python web applications.

Gunicorn is often used as a production-level server for hosting Python web applications,
including Django and Flask applications. It acts as a middleman between the web application and 
the web server, handling incoming HTTP requests, executing the application code, and returning the 
HTTP responses.

Here are some key features and characteristics of Gunicorn:

Concurrency Model: Gunicorn uses a pre-fork worker model, where the master process forks multiple
    worker processes to handle incoming requests concurrently. Each worker process can handle
    multiple requests in parallel, allowing for efficient utilization of system resources.

Scalability: Gunicorn is designed to scale horizontally by spawning multiple worker processes. 
    It can handle a large number of concurrent connections and distribute the incoming requests 
    among the worker processes.

WSGI Compatibility: Gunicorn adheres to the WSGI specification, which is a standard interface between
    web servers and web applications in Python. This compatibility allows Gunicorn to work seamlessly
    with various Python web frameworks and applications that support the WSGI standard.

Configuration Options: Gunicorn provides a range of configuration options, allowing customization of 
    various server settings, such as the number of worker processes, timeout values, logging, and 
    more. The configuration can be specified in command-line arguments or in a configuration file.
    
Integration with Deployment Tools: Gunicorn integrates well with popular deployment tools and
    frameworks, such as Nginx, Apache, Docker, and various orchestration tools. It can be used as
    part of a larger deployment stack to ensure reliable and efficient hosting of Python web 
    applications.

Gunicorn simplifies the process of deploying Python web applications, providing a robust and 
efficient server for running WSGI-compatible applications. It offers a balance between simplicity
and performance, making it a popular choice for hosting Python web applications in production 
environments.

21. nginx
22. django password encryption hashing algorithm?
ans. Django's implementation of PBKDF2 with HMAC-SHA256.

23. ldef security

24. class, static , instanse method and variable
25. select related and prefetch related?
ans. In Django, both select_related() and prefetch_related() are query optimization methods that can
improve performance when working with related objects in a database.

select_related():

i)The select_related() method is used to fetch related objects in a single database query by performing
a SQL join.
ii)It works for ForeignKey and OneToOneField relationships and follows the relationship between models 
to retrieve the related objects.
iii)By using select_related(), you can reduce the number of database queries and improve performance 
when accessing related objects.
Example:
from myapp.models import Author

# Retrieve authors and their associated books with a single database query
authors = Author.objects.select_related('book')

for author in authors:
    print(author.name, author.book.title)

prefetch_related():

i)The prefetch_related() method is used to retrieve related objects in separate database queries,
but in a more efficient way compared to making individual queries.
ii)It works for many-to-many and many-to-one relationships and fetches the related objects using a 
separate query per relationship.
iii) By using prefetch_related(), you can reduce the number of database queries and improve
performance when accessing related objects.
Example:
from myapp.models import Category

# Retrieve categories and their associated products with a single database query
categories = Category.objects.prefetch_related('products')

for category in categories:
    print(category.name)
    for product in category.products.all():
        print(product.name)

In summary, select_related() retrieves related objects using a SQL join in a single query,
while prefetch_related() retrieves related objects in separate queries but in a more efficient manner
compared to making individual queries. The choice between them depends on the specific relationship 
type, the number of related objects, and the specific use case to optimize the performance of your 
Django queries.
        
    
26. get(), filter() and  all() methods difference?
27. map(), filter() and reducer() method and example
ans. 
In Python, map(), filter(), and reduce() are built-in higher-order functions that operate on
iterables (such as lists, tuples, or sets) and perform specific operations on their elements.
def square(x):
    return x * x
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(square, numbers)
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]


def is_even(x):
    return x % 2 == 0
numbers = [1, 2, 3, 4, 5]
even_numbers = filter(is_even, numbers)
print(list(even_numbers))  # Output: [2, 4]


from functools import reduce
def multiply(x, y):
    return x * y
numbers = [1, 2, 3, 4, 5]
product = reduce(multiply, numbers)
print(product)  # Output: 120 (1 * 2 * 3 * 4 * 5)


28. middilware, custom middilware
ans. Middleware is something that executes between the request and response. In simple words,
you can say it acts as a bridge between the request and response. Similarly In Django when a request 
is made it moves through middlewares to views and data is passed through middleware as a response. 


29. decorater and custom decorater
30. mixins
ans. Mixin is a type of multiple inheritances wherein you can combine behaviors and attributes of 
more than one parent class. It provides us with an excellent way to reuse code from multiple classes.
One drawback of using these mixins is that it becomes difficult to analyze what a class is doing and 
which methods to override in case of its code being too scattered between multiple classes.



31. modelviewset and apiview difference
32. generator?
33. Iterator and iteration?
34. enumerator?
ans. In Python, the enumerate() function is a built-in function that allows you to iterate over an 
iterable (such as a list, tuple, or string) while also tracking the index or position of each item 
in the iterable. It returns an iterator of tuples containing the index and value of each element.
The basic syntax of the enumerate() function is as follows:
enumerate(iterable, start=0)
fruits = ['apple', 'banana', 'orange']

for index, fruit in enumerate(fruits):
    print(index, fruit)
output:-
0 apple
1 banana
2 orange

By default, enumerate() starts the index from 0, but you can specify a different starting value 
if needed. For example:

fruits = ['apple', 'banana', 'orange']

for index, fruit in enumerate(fruits, start=1):
    print(index, fruit)

output:-
1 apple
2 banana
3 orange



35. zip()
ans. In Python, the zip() function is a built-in function that takes iterables (such as lists,
tuples, or strings) as input and returns an iterator of tuples where each tuple contains the 
corresponding elements from the input iterables.
The zip() function combines elements from multiple iterables based on their positions and stops when
the shortest iterable is exhausted. In other words, it aligns the elements pairwise.

Here's the basic syntax of the zip() function:
numbers = [1, 2, 3]
letters = ['a', 'b', 'c']
names = ['John', 'Jane', 'Jim']

zipped = zip(numbers, letters, names)

for item in zipped:
    print(item)

output:-
(1, 'a', 'John')
(2, 'b', 'Jane')
(3, 'c', 'Jim')

zipped_list = list(zip(numbers, letters, names))
print(zipped_list)
output:-
[(1, 'a', 'John'), (2, 'b', 'Jane'), (3, 'c', 'Jim')]

The zip() function is commonly used when you need to iterate over multiple lists or iterables in
parallel and process their elements together.

36. pickling and unpicking
ans. Pickling and unpickling are processes in Python that allow you to convert objects into a byte 
stream (serialization) and restore the objects from the byte stream (deserialization), respectively. 
This process is useful for storing or transmitting objects across different systems or for preserving
object state between program runs.

Pickling:
Pickling is the process of converting a Python object into a byte stream, which can be stored in a 
file or transmitted over a network.
The pickle module in Python provides the dump() and dumps() functions to serialize objects into a 
byte stream.
The dump() function is used to serialize an object and write it to a file, while dumps() returns a 
serialized byte stream as a string.
Example:
import pickle

data = {'name': 'John', 'age': 30}

# Serialize and write the object to a file
with open('data.pickle', 'wb') as file:
    pickle.dump(data, file)

Unpickling:
Unpickling is the process of restoring a Python object from a byte stream back into its original form.
The pickle module provides the load() and loads() functions to deserialize objects from a byte stream.
The load() function is used to read and deserialize an object from a file, while loads() takes a 
serialized byte stream as input and returns the deserialized object.
Example:
import pickle

# Deserialize the object from the file
with open('data.pickle', 'rb') as file:
    data = pickle.load(file)
    
print(data)  # Output: {'name': 'John', 'age': 30}
Pickling and unpickling allow you to preserve complex object structures, including custom classes, 
nested data structures, and more. However, it's important to note that the pickle module is specific
to Python and not compatible with other programming languages.

When pickling and unpickling objects, ensure that you trust the source of the byte stream, as 
unpickling untrusted data from an unauthenticated source can pose security risks. It is recommended 
to use pickling within trusted environments or with data you have full control over.

37. jwt authentication
38. celery
Celery is a distributed task queue system for Python that enables you to execute tasks asynchronously
in a distributed manner. It is commonly used for handling time-consuming or resource-intensive tasks
in web applications or backend systems.

Key features of Celery include:
Distributed Architecture: Celery allows you to distribute tasks across multiple workers 
(also known as Celery workers) for parallel processing. Workers can run on different machines or 
processes, enabling scalability and efficient resource utilization.

Task Queues: Celery uses message brokers, such as RabbitMQ, Redis, or others, as the intermediary 
for communication between the client (task producer) and workers (task consumers). Tasks are placed
in a queue and workers pick up tasks from the queue for execution.

Task Routing and Priority: Celery provides flexible task routing options, allowing you to specify 
which workers should handle specific tasks based on criteria such as task type, priority, or custom
routing rules.

Task Result Tracking: Celery tracks the status and results of tasks, allowing you to check the 
progress of tasks and retrieve their results asynchronously.

Task Scheduling: Celery includes built-in support for task scheduling, allowing you to define
periodic tasks (scheduled to run at specific intervals) or cron-like tasks
(scheduled based on specific time patterns).

Error Handling and Retry: Celery provides mechanisms for handling task failures, including automatic
retries, task timeouts, and error handling options to ensure reliable task execution.

Celery integrates well with various web frameworks and systems, including Django, Flask, and others.
It enables you to offload time-consuming or resource-intensive tasks (such as image processing, 
data processing, or sending emails) to background workers, freeing up the main application to handle
user requests more efficiently.

To use Celery, you need to set up a Celery instance, define tasks using Python functions or classes 
decorated with @task, configure the Celery worker(s) to connect to the message broker, and start the
worker(s) to process the tasks.

Celery provides extensive documentation and offers various configuration options and extensions to
customize its behavior based on your specific requirements.

Official Celery documentation: https://docs.celeryproject.org/

39. orm
ans. In Django, the most notable feature is Object-Relational Mapper (ORM), which allows you to
interact with app data from various relational databases such as SQLite, MySQL, and PostgreSQL. 
Django ORM is the abstraction between web application data structure (models) and the database where
the data is stored. Without writing any code, you can retrieve, delete, save, and perform other
operations over the database.
The main advantage of ORMs is rapid development. ORMs make projects more portable. It's easier to 
change the database with Django ORM.

40. Signals and custom signals
ans. Whenever there is a modification in a model, we may need to trigger some actions. 
Django provides an elegant way to handle these in the form of signals. The signals are the utilities 
that allow us to associate events with actions. We can implement these by developing a function that
will run when a signal calls it.


41. what is csrf token
42. How can you combine multiple QuerySets in a View?
ans. Initially, Concatenating QuerySets into lists is believed to be the easiest approach. 
Here’s an example of how to do that:
from itertools import chain
result_list = list(chain(model1_list, model2_list, model3_list))

43. Explain Q objects in Django ORM?
Q objects are used to write complex queries, as in filter() functions just `AND` the conditions
while if you want to `OR` the conditions you can use Q objects. Let’s see an example:

from django.db import models
from django.db.models import Q
>> objects = Models.objects.get(
   Q(tag__startswith='Human'),
   Q(category=’Eyes’) | Q(category=’Nose’)
)
```Query Executed
SELECT * FROM Model WHERE tag LIKE ‘Human%’ AND (category=’Eyes’ OR category=’Nose’)
```
44. abstract base class in python?
In Python, an abstract base class (ABC) is a class that cannot be instantiated and is meant to serve
as a blueprint for other classes. It defines a common interface for its subclasses, specifying a set
of methods that must be implemented by any class that inherits from it.
The abc module in Python's standard library provides the infrastructure for defining abstract base 
classes. To create an abstract base class, you need to inherit from the ABC class and use the 
@abstractmethod decorator to mark the methods that must be implemented by the subclasses.
Here's an example of how to define an abstract base class in Python:
from abc import ABC, abstractmethod

class MyABC(ABC):

    @abstractmethod
    def my_method(self):
        pass

    @abstractmethod
    def another_method(self, arg):
        pass

Any class that inherits from MyABC must implement these two methods.

If a subclass does not implement one or more of the abstract methods defined in the base class, 
it will raise a TypeError at runtime.
Here's an example of a subclass that implements the abstract methods:
class MyClass(MyABC):

    def my_method(self):
        # Implementation of my_method
        pass

    def another_method(self, arg):
        # Implementation of another_method
        pass
Using abstract base classes allows you to define common interfaces and enforce certain method 
implementations across multiple classes, promoting code reusability and providing a clear structure
for code.
````

````python
What is a microservices?

Microservices, also known as the microservices architecture, is a software development approach in
which a single application is built as a collection of small, loosely coupled, and independently 
deployable services. Each service represents a specific business capability and can be developed, 
deployed, and scaled independently of the others. These services communicate with each other over a 
network, typically using APIs (Application Programming Interfaces).

Create the User Service:
Create a file named user_service.py:
    
from flask import Flask, jsonify

app = Flask(__name__)

users = {
    1: {"id": 1, "name": "Alice"},
    2: {"id": 2, "name": "Bob"}
}

@app.route('/users/<int:user_id>')
def get_user(user_id):
    user = users.get(user_id)
    if user:
        return jsonify(user)
    else:
        return jsonify({"error": "User not found"}), 404

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5001)
    
2. Create the Order Service:
Create a file named order_service.py:

from flask import Flask, jsonify

app = Flask(__name__)

orders = {
    1: {"id": 1, "product": "Product A"},
    2: {"id": 2, "product": "Product B"}
}

@app.route('/orders/<int:order_id>')
def get_order(order_id):
    order = orders.get(order_id)
    if order:
        return jsonify(order)
    else:
        return jsonify({"error": "Order not found"}), 404

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5002)

    
Run the Services:
Open two terminal windows and run the following commands:
python user_service.py
python order_service.py


Now you have two simple microservices running. The user service listens on port 5001, and 
the order service listens on port 5002. You can access these services by sending HTTP requests to 
the respective endpoints, 
e.g. http://localhost:5001/users/1 and 
http://localhost:5002/orders/1.
````

````python
def min_jumps_to_reach_end(arr):
    n = len(arr)

    if n == 0 or arr[0] == 0:
        return -1

    jumps = 1  # Number of jumps needed to reach the end
    max_reach = arr[0]  # Maximum index that can be reached with the current number of jumps
    steps_available = arr[0]  # Number of steps available with the current jump

    for i in range(1, n):
        if i == n - 1:
            return jumps

        max_reach = max(max_reach, i + arr[i])
        steps_available -= 1

        if steps_available == 0:
            jumps += 1

            if i >= max_reach:
                return -1  # Cannot reach the end

            steps_available = max_reach - i

    return -1  # If the loop ends without reaching the end

# Example usage:
arr = [10,14,29,21,17,4,18,20,18,22,21,14,27,12,3,28,17,0,2,18,8,20,26,16,9,23,25,20,7,27,5,7,16,5,25,11,3,7,2,17,14,6,12,14,23,25,26,5,18,1,6,10,9,12,2,25,29,12,19,4,8,5,8,30,2,22,24,30,7,24,8,15,16,2,11,20]
result = min_jumps_to_reach_end(arr)
print(result)
````
