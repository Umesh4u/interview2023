```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
class LinkedList:
    def __init__(self):
        self.head=None
        
    def insertAtBegining(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node
     
    def insertAtEnd(self,data):
        new_node=Node(data)
        if self.head is None:
            self.head=new_node
            return 
        temp=self.head
        while temp.next is not None:
            temp=temp.next
        temp.next=new_node
        
    def insertInBetween(self,data,position):
        new_node=Node(data)
        temp=self.head
        if position==0:
            new_node.next=temp
            temp=new_node
            return 
        
        for i in range(position-1):
            temp=temp.next
        new_node.next=temp.next
        temp.next=new_node
        
    def PrintData(self):
        temp=self.head
        while temp is not None:
            print(temp.data, end= " ")
            temp=temp.next
        
    def deleteNode(self,position):
        temp=self.head
        if position==0:
            head=temp.next
            return
        for _ in range(position-2):
            temp=temp.next
            
        temp.next=temp.next.next
        
        
    def deleteNodeValueWise(self,data):
        temp=self.head
        if data==temp.data: # if value is first node then this condition will be helpfull
            head=temp.next
            return 
        
        while temp.next is not None: #here seraching will start from 2nd node because temp.next 
            if temp.next.data==data:
                temp.next=temp.next.next
            temp=temp.next
            
    def reverseLL(self):
        temp=self.head
        prev=None
        curr=temp
        while curr is not None:
            nextt=curr.next
            curr.next=prev
            prev=curr
            curr=nextt
        self.head=prev
#         return sel
        
    def checkPalindrome(self):
        #calculate length of linked list
        n=0
        temp=self.head
        while temp is not None:
            temp=temp.next
            n+=1
        length = n//2

        #break the LL from mid
        temp=self.head
        for _ in range(length-1):
            temp=temp.next
        
        head2 =temp.next
        temp.next=None

        #reverse 2nd half LL
        prev = None
        curr = head2
        while curr is not None:
            nextt=curr.next
            curr.next=prev
            prev=curr
            curr=nextt
        head2=prev

        #compare both halfs
        t1=self.head 
        t2=head2
        while t1 is not None and t2 is not None:
            if t1.data!=t2.data:
                return "not palindrome"
            t1=t1.next
            t2=t2.next
        return "Palindrome"
```
